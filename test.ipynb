```python
import pandas as pd
from ast import literal_eval

df = pd.read_csv("expenses.csv")

columns = df.columns

for col in columns:
    if col.__contains__("Unnamed"):
        print("This is the column: ", col)

        df.drop(col, axis = 1, inplace=True)

df.index = ["income", "expenses", "total_expenses", "cashouts", "savings"]

df
```

```python

df.drop("October", axis=1, inplace=True)
df.drop("November", axis=1, inplace=True)
df.drop("December", axis=1, inplace=True)

df
```

```python
df.to_csv("expenses.csv")
```
